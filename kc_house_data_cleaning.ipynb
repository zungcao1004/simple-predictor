```python
import pandas as pd
```

```python
df = pd.read_csv(filepath_or_buffer='kc_house_data.csv')
df.info()
```

```python
cols = ['id', 'date', 'sqft_lot', 'waterfront', 'view', 'sqft_basement', 'sqft_above', 'yr_built', 'yr_renovated', 'zipcode', 'lat', 'long', 'sqft_living15', 'sqft_lot15']
df = df.drop(cols, axis=1)
```

```python
df['bathrooms'] = df['bathrooms'].apply(lambda x: int(float(x)) if float(x).is_integer() else int(x))
df['floors'] = df['floors'].apply(lambda x: int(float(x)) if float(x).is_integer() else int(x))
df
```

```python
df.to_csv('kc_house_data_cleaned.csv', index=False)
```
